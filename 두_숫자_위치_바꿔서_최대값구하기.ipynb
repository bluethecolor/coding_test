- 숫자열이 주어지고, 임의의 위치의 두 숫자를 바꿀 수 있음.
- 같은 자리의 숫자를 바꾸는건 안되고, 한번 바꾼 자리를 중복해서 바꿀 수 있음
- 이렇게 해서 반복 횟수를 전부 채웠을 때 최대값 구하기

```python
# que = [1,2,3] #1  321
# que = [2,7,3,7] #1   7732
# que = [7,5,7,1,4,8] #1   857147
# que = [7,8,4,6,6] #2    87664
# que = [3,2,8,8,8] #2    88832
# que = [7,7,7,7,7,0] #5   777770
# que = [4,3,6,6,5,9] #2    966354
# que = [4,3,1,1,5,9] #7    954311
# que = [1,1,2,2,3,3] #3    332211
que = [4,5,6,7,8,9] #10   987645

reap = 10
```

```python
T = int(input())
for test_case in range(1, T + 1):

    que, reap = map(int, input().split()) 

    # que를 123 -> [1,2,3] 형태로 바꿔줌.
    que = str(que)
    que = list(que)
    que = list(map(int, que)) 


    reap_time = 0

    for i in range(len(que)):   # i뒤에 있는 값들 중 i보다 큰 요소들의 리스트를 생성
        list_que = []
        for k in range(i+1, len(que)):
            if que[i] < que[k]:
                list_que.append(que[k])
        list_que.append(que[i]) # 자기 자신도 넣어주는 이유는 if max(list_que) == que[i]: 과정 떄문에. 자기가 최대값이면 넘겨야됨;;
        
        
        if i == len(que)-1 and reap_time < reap :  # 위에꺼 다 돌렸는데도 reap_time이 안채워지면 진행

            if len(set(que)) < len(que):  # que 안에 중복값이 있으면 그거끼리 막 바꿔주면 됨으로 그대로 출력
                
                print(f"{test_case} {int(''.join(str(x) for x in que))}")
                break
            else:
                for r in range(reap - reap_time):   # 중복값이 없으면 제일 뒤에 두 값을 남은 횟수만큼 돌려줘야 최대값 뽑음.
                    que[len(que) -2] , que[len(que) -1] = que[len(que) -1], que[len(que) -2]

                print(f"{test_case} {int(''.join(str(x) for x in que))}")
                break


        if max(list_que) == que[i]:  # 최대값이랑 같으면 그냥 그대로 둬도 됨으로 패스
            continue

        elif max(list_que) != que[i]: 
            
            if list_que.count(max(list_que)) > 1: # 최대값이 2개 이상인 경우

                rest_list = list(filter(lambda x: que[x] == max(list_que), range(len(que)))) # list_que안의 최대값의 que에서의 인덱스값 전부 뽑아줌.
                list_vs = []  # 밑에서 만들어진 것들끼리 비교하기 위해 여기에 모아줌.
            
                for z in rest_list: # rest_list 안의 최대값의 que에서의 인덱스에 해당하는 값을 que의 i번째 값과 변경해주는 과정
                    que_3 = que.copy()

                    if que_3[i+1] < que_3[i] and que_3.count(max(que_3)) > 1 and reap - reap_time >= 2: # 32888같은 예외 사항 대비

                        M = que_3.count(max(que_3))
                        A = que_3.index(max(que_3)) + M - 2
                        B = que_3.index(max(que_3)) + M - 1

                        que_3[i],que_3[i+1], que_3[A] , que_3[B] = que_3[A] , que_3[B], que_3[i],que_3[i+1]
                        reap_time += 1  # 한번에 두 개 바꾸니 1 추가로 증가

                    else:  # 위 예외 사항이 아닌 경우 이렇게

                        if z !=0:  # 최대값이 que[0] 자리에 있는 경우는 제외해야;;
                            que_3[i] , que_3[z] = que_3[z],que_3[i]

                    que_4 = que_3.copy()
                    max_que = int(''.join(str(x) for x in que_4)) # [1,2,3] => 123으로 변경해줌.
                    list_vs.append(max_que)
                dop = max(list_vs)

                que = list(map(int, str(dop))) # 123 -> [1,2,3]으로 다시 변경해줌.
                reap_time += 1

            else: # 최대값이 1개만 있는 경우

                L = que[i+1:].index(max(list_que)) + i + 1 # i앞에 최대값이 포함되면 안되고, i는 앞에서 최대값 아닌게 확인되야 오는 단계이므로 que[i+1:], 그리고 i+1만큼 더해줘야 인덱스 위치 맞춰짐.
                que[i] , que[L] = que[L], que[i]
                reap_time += 1
        
        if reap_time >= reap:
            print(f"{test_case} {int(''.join(str(x) for x in que))}")
            break
```
